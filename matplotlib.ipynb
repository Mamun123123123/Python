```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.arange(1,11)
y = x**2
plt.plot(x,y)
plt.title("simple plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```
